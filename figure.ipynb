```python
dev_x=[25,26,27,29,30]
dev_y=[584696,467000,466752,496570,563200]
#if we have the same y we write here sharey=True
#we can change dimensions by figure=(x,y)
#if we use 2Diminsion we should write axes [i,j]
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].set_title('salary for each age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Salary')
axes[0].plot(dev_x,dev_y,color='b',linestyle='--',label='Mean Dev Salary')

py_dev_x=[25,26,27,29,30,31,32]
py_dev_y=[567543,472020,467352,695672,573203,738758,967327]
axes[1].plot(py_dev_x,py_dev_y,color='r',label='Mean Python Salary')
axes[1].set_title('Python developer salary for each age')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Python')


axes[0].legend()
axes[1].legend()
fig.savefig('salaryage.png')
plt.tight_layout()
plt.show()
```
